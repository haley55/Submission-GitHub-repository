```python
import pandas as pd
import numpy as np
```

```python
df_download_speed= pd.read_csv("plot_hackathon\Download speed for all the counties.csv")
```

```python
df_download_speed.head()
```

```python
df_download_speed.describe()
```

```python
df_download_speed.fillna((df_download_speed.mean()), inplace=True) #fill NaN result with mean of that column
```

```python
df_download_speed.head()
```

```python
Downloadspeed_ave_data = df_download_speed.copy()
Downloadspeed_ave_data['Downloadspeed_average'] = Downloadspeed_ave_data.mean(numeric_only=True, axis=1)
Downloadspeed_ave_data.head()
```

```python
Downloadspeed_ave_data["Unnamed: 0"]
```

```python
Downloadspeed_ave_data.rename(columns={'Unnamed: 0': 'County'}, inplace=True)
Downloadspeed_ave_data.head()
```

```python
df_speed=Downloadspeed_ave_data[["County","Downloadspeed_average"]]
df_speed.head()
```

```python
df_income= pd.read_csv("plot_hackathon\download_allfamilyincome_all.csv")
df_income.head()
```

```python
N = len(df_income.index)
N
```

```python
import itertools
df_income['grp'] = list(itertools.chain.from_iterable([x]*4 for x in range(0, N//4)))
df_income.head()
```

```python
df_income_ave=df_income.groupby('grp').mean()
df_income_ave
```

```python
df_county=pd.DataFrame(df_income['CSD'].unique())
```

```python
df_county.head()
```

```python
df_county.rename(columns={'0': 'County'}, inplace=True)
df_county.head()
```

```python
df_county['Income']=df_income_ave['OriginalValue']
df_county.head()
```

```python
for col in df_county.columns:
    print(col)
```

```python
df_county.rename(columns={'0': 'County'}, inplace=True)
df_county.head()
```

```python
df_county.columns=['Partname_County','Income']
df_county.head()
```

```python
pat = "|".join(df_speed.County)
df_county.insert(0, 'County',df_county['Partname_County'].str.extract("(" + pat + ')', expand=False))
print (df_county)
  
```

```python
df_county.info()
```

```python
df_county.head(90)
```

```python
result = pd.merge(df_speed,
                 df_county[['County', 'Income']],
                 on='County')
result.head()
```

```python
result.tail()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = result
sns.set(rc={'figure.figsize':(11.7,8.27)})
b1=sns.lineplot(data=df['Downloadspeed_average'], color="r")
ax2 = plt.twinx()
b2=sns.lineplot(data=df['Income'], color="b", ax=ax2)
b1.set_xticks(range(len(df)))
b1.set_xticklabels(df['County'],rotation=90)
plt.savefig('Average Download speed_incove_vs_34county.png',dpi=300, bbox_inches='tight', pad_inches=0)
```
